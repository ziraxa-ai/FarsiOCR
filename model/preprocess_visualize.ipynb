```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path
import random, math
import pandas as pd
repo = Path.cwd().parents[0]
```

```python
df = pd.read_csv(repo /"Dataset/Create Dataset/train.csv")
labels = df["label"].unique()
n = min(8, len(labels))
fig, axes = plt.subplots(n, 8, figsize=(12, n*1.5))

for i,l in enumerate(sorted(labels)[:n]):
    sub = df[df["label"] == l].sample(min(8, len(df)//len(labels)))

    for j, (idx, row) in enumerate(sub.iterrows()):
        img = Image.open(repo/"Dataset/Create Dataset"/row['image_path']).convert("L")
        axes[i, j].imshow(img, cmap="gray")
        axes[i, j].axis("off")

        if j == 0:axes[i, j].set_title(l)
    plt.tight_layout()
    plt.show()
```
